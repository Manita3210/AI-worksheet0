• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]])

1. Prove A.A-1 = I.





```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
A_inverse = np.linalg.inv(A)
i = np.dot(A, A_inverse)
print("A · A⁻¹ =\n", i)
```

2. Prove AB ̸= BA.

```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
AB = np.dot(A, B)
BA = np.dot(B, A)
print("A·B =\n", AB)
print("B·A =\n", BA)
print("A·B == B·A ?", np.allclose(AB, BA))
```

3. Prove (AB)^T =B^T*A^T

```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
AB_T = np.dot(A, B).T
B_T_A_T = np.dot(B.T, A.T)
print("(AB)ᵀ =\n", AB_T)
print("Bᵀ·Aᵀ =\n", B_T_A_T)
print("Are they equal?", np.allclose(AB_T, B_T_A_T))
```

Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1

x − y + 2z = −3

3x + y − z = 9


```python
import numpy as np
coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])
constants = np.array([-1, -3, 9])
solution = np.dot(np.linalg.inv(coeff_matrix), constants)
print("Solution [x, y, z] =", solution)
```
